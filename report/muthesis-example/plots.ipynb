# Ticket Count Plot

```python
import matplotlib.pyplot as plt

x = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
y = [3958, 5475, 6719, 8598, 9775, 9776, 12586, 14964, 15921]

plt.figure(figsize=(6.5, 3.5))
plt.plot(x, y, marker='x')
plt.title('Number of CSD3 Helpdesk Tickets per Year (2017-2025)')
plt.xlabel('Year')
plt.ylabel('Number of Tickets')
plt.grid()
plt.xticks(x)
plt.savefig("graphs/ticket_counts.pdf", bbox_inches="tight")
plt.show()
```
